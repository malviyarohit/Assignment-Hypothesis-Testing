# 📊 HYPOTHESIS TESTING ASSIGNMENT  
**Complete Questions & Answers**  

---

## QUESTION 1  
What is a null hypothesis (H₀) and why is it important in hypothesis testing?

**Answer:**  
The null hypothesis (H₀) is a statement of “no effect” or “no difference”.  
Example: H₀: μ = 10 minutes.  
It is important because it is the default assumption we try to disprove with evidence. All hypothesis testing starts by assuming H₀ is true.

---

## QUESTION 2  
What does the significance level (α) represent in hypothesis testing?

**Answer:**  
α is the probability of rejecting the null hypothesis when it is actually true (Type I error rate).  
Common values: 0.05 (5%) or 0.01 (1%).  
It is the maximum risk we are willing to take of making a wrong decision.

---

## QUESTION 3  
Differentiate between Type I and Type II errors.

**Answer:**  
- **Type I error** (α): Rejecting H₀ when it is true (False Positive).  
  Example: Saying the medicine works when it actually doesn’t.  
- **Type II error** (β): Failing to reject H₀ when it is false (False Negative).  
  Example: Saying the medicine doesn’t work when it actually does.

---

## QUESTION 4  
Explain the difference between a one-tailed and two-tailed test. Give an example of each.

**Answer:**  
- **One-tailed test**: Tests direction (greater than or less than).  
  Example: H₁: μ > 10 (training improves productivity).  
- **Two-tailed test**: Tests any difference (not equal).  
  Example: H₁: μ ≠ 10 (average time is different from 10 minutes).

---

## QUESTION 5  
A company claims that the average time to resolve a customer complaint is 10 minutes. A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At α = 0.05, test the claim.

**Answer:**  
**Step-by-step solution:**  
H₀: μ = 10 minutes  
H₁: μ ≠ 10 minutes (two-tailed test)  

t-statistic = (x̄ - μ) / (s / √n) = (12 - 10) / (3 / √9) = 2 / 1 = 2  
df = 9 - 1 = 8  

Critical t-value (α = 0.05, two-tailed, df=8) = ±2.306  
Since |2| < 2.306, **fail to reject H₀**.  

Conclusion: There is not enough evidence to reject the company’s claim at 5% significance level.

**EXCEL VERSION (Exact Steps):**  
1. Open Excel  
2. In cell A1: 12 (in B1: 10) (in C1: 3) (in D1: 9)  
3. In E1 type: `=(A1-B1)/(C1/SQRT(D1))` → t-statistic = 2  
4. In F1 type: `=T.INV.2T(0.05,8)` → critical value = 2.306  
5. In G1 type: `=IF(ABS(E1)>F1,"Reject H0","Fail to reject H0")` → Fail to reject H0

## QUESTION 6  
When should you use a Z-test instead of a t-test?

**Answer:**  
Use Z-test when:  
- Population standard deviation (σ) is known, OR  
- Sample size n ≥ 30 (Central Limit Theorem applies), OR  
- Data is normally distributed and σ is known.  

Use t-test when σ is unknown and n is small (<30).

## QUESTION 7  
The productivity of 6 employees was measured before and after a training program.  
(Use the table given)  
At α = 0.05, test if the training improved productivity.

**Answer:**  
This is a **paired t-test** (before-after same employees).  

Differences (After - Before): 5, 5, 1, 3, 1, 3  
Mean difference (d̄) = 3  
Standard deviation of differences (s_d) ≈ 1.789  

t-statistic = d̄ / (s_d / √n) = 3 / (1.789 / √6) ≈ 4.11  
df = 6 - 1 = 5  

Critical t-value (one-tailed, α = 0.05, df=5) = 2.015  
Since 4.11 > 2.015, **reject H₀**.  

Conclusion: The training program significantly improved productivity at 5% level.

**EXCEL VERSION (Exact Steps – Easiest Way):**  
1. Enter data:  
   Column A (Before): 50,60,58,55,62,56  
   Column B (After): 55,65,59,58,63,59  
2. In any empty cell type: `=T.TEST(A1:A6,B1:B6,1,1)`  
   → Result ≈ 0.003 (p-value)  
3. Since p-value 0.003 < 0.05 → Reject H₀ (training improved productivity)

## QUESTION 8  
A company wants to test if product preference is independent of gender.  
(Use the contingency table given)  
At α = 0.05, test independence.

**Answer:**  
This is a **Chi-square test for independence**.  

Expected frequencies:  
Male-A: 20, Male-B: 30  
Female-A: 20, Female-B: 30  

Chi-square statistic = Σ(O - E)²/E = 16.667  
df = (2-1)(2-1) = 1  

Critical value (α = 0.05, df=1) = 3.841  
Since 16.667 > 3.841, **reject H₀**.  

Conclusion: Product preference is **not independent** of gender (there is association).

**EXCEL VERSION (Exact Steps):**  
1. Create observed table in Excel:  
   A2:B3 → Male: 30, 20  
   A4:B5 → Female: 10, 40  
2. Calculate expected (or let Excel do it):  
   In C2: `=50*40/100` (=20) etc. for all 4 cells  
3. In any cell type: `=CHISQ.TEST(A2:B3,C2:D3)`  
   → p-value ≈ 0.00004  
4. Since p < 0.05 → Reject H₀
